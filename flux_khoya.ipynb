```python
import threading
import time

def keep_alive():
    while True:
        time.sleep(300)  # Sleep for 5 minutes
        print("Keeping session alive...")

# Start the thread
thread = threading.Thread(target=keep_alive)
thread.daemon = True
thread.start()
```
